```python
import pandas as pd
```

```python
import json
```

```python
def coldetails(y, name, sorting_key='column_position'):
    det = y[name]
    colum = sorted(det, key=lambda x : x[sorting_key] )
    return [x['column_name'] for x in colum]
```

```python
ehek = json.load(open('data/retail_db/schemas.json'))
```

```python
turu = coldetails(ehek , 'orders')
```

```python
orders = pd.read_csv('data/retail_db/orders/part-00000', names = turu)
```
